```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_profile_df = pd.read_csv('userprofile.csv')
user_cuisine_df = pd.read_csv('usercuisine.csv')
```

```python
user_profile_df = user_profile_df[['userID', 'smoker', 'budget','drink_level']]
```

```python
user_profile_df['smoker'] = user_profile_df['smoker'].replace(to_replace='?', value=user_profile_df['smoker'].mode()[0])
```

```python
user_profile_df['budget'] = user_profile_df['budget'].replace(to_replace='?', value=user_profile_df['budget'].mode()[0])
```

```python
user_profile_df['drink_level'].unique()
```

```python
user_cuisine_df
```

```python
user_profile_df['smoker'] = user_profile_df['smoker'].replace(to_replace='false', value=0)
user_profile_df['smoker'] = user_profile_df['smoker'].replace(to_replace='true', value=1)
```

```python
user_profile_df['budget'] = user_profile_df['budget'].replace(to_replace='low', value=0)
user_profile_df['budget'] = user_profile_df['budget'].replace(to_replace='medium', value=1)
user_profile_df['budget'] = user_profile_df['budget'].replace(to_replace='high', value=2)
```

```python
user_profile_df['drink_level'] = user_profile_df['drink_level'].replace(to_replace='abstemious', value=0)
user_profile_df['drink_level'] = user_profile_df['drink_level'].replace(to_replace='social drinker', value=1)
user_profile_df['drink_level'] = user_profile_df['drink_level'].replace(to_replace='casual drinker', value=2)
```

```python
user_profile_df
```

```python
user_profile_similarity = cosine_similarity(user_profile_df[['smoker', 'budget','drink_level']])
user_profile_similarity_df = pd.DataFrame(user_profile_similarity, index=user_profile_df['userID'], columns=user_profile_df['userID'])
```

```python
user_profile_similarity_df
```

```python
user_cuisine_matrix = user_cuisine_df.pivot_table(index='userID', columns='Rcuisine', aggfunc='size', fill_value=0)
```

```python
user_cuisine_matrix
```

```python
user_cuisine_similarity = cosine_similarity(user_cuisine_matrix)
user_cuisine_similarity_df = pd.DataFrame(user_cuisine_similarity, index=user_cuisine_matrix.index, columns=user_cuisine_matrix.index)
```

```python
user_cuisine_similarity_df
```

```python
cuisine_weight = 0.8
profile_weight = 0.2
```

```python
user_similarity_df = cuisine_weight*user_cuisine_similarity_df + profile_weight*user_profile_similarity_df
```

```python
user_similarity_df
```

```python
def get_top_k_cuisines_for_user(user_id, k=5):
    # Find similar users to the target user
    similar_users = user_similarity_df[user_id].sort_values(ascending=False)[1:k+1].index

    # Get the cuisines preferred by similar users
    top_k_cuisines = user_cuisine_matrix.loc[similar_users].sum().sort_values(ascending=False)[:k].index

    return top_k_cuisines
```

```python
user_id_to_recommend = 'U1016'  # Replace with the user ID you want to recommend cuisines for
top_k_cuisines = get_top_k_cuisines_for_user(user_id_to_recommend, k=5)
print(f"Top-K cuisines for user {user_id_to_recommend}:")
print(top_k_cuisines)
```
