```python
import pandas as pd
import numpy as np
```

```python
user_profile_df = pd.read_csv('userprofile.csv')
user_cuisine_df = pd.read_csv('usercuisine.csv')
```

```python
user_profile_df_temp = user_profile_df.drop(['latitude','longitude','weight','height','birth_year'],axis=1)
```

```python
for column in user_profile_df_temp.columns:
  user_profile_df_temp[column] = user_profile_df_temp[column].replace(to_replace='?', value=user_profile_df_temp[column].mode()[0])
```

```python
def jaccard_similarity(set1, set2):
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union
```

```python
def jaccard_similarity_matrix(data):
    num_items = len(data)
    similarity_matrix = [[0] * num_items for _ in range(num_items)]

    for i in range(num_items):
        for j in range(num_items):
            if i == j:
                similarity_matrix[i][j] = 1.0
            else:
                similarity_matrix[i][j] = jaccard_similarity(set(data[i]), set(data[j]))

    return similarity_matrix
```

```python
jaccard_similarity_matrix = jaccard_similarity_matrix(np.array(user_profile_df_temp.drop(['userID'],axis=1)))
```

```python
user_profile_similarity_df = pd.DataFrame(jaccard_similarity_matrix, index=user_profile_df_temp['userID'], columns=user_profile_df_temp['userID'])
```

```python
user_profile_df_body = user_profile_df[['userID','weight','height']]
```

```python
def manhattan_distance(vec1, vec2):
    return sum(abs(a - b) for a, b in zip(vec1, vec2))
```

```python
def manhattan_similarity_matrix(data):
    num_points = len(data)
    similarity_matrix = [[0] * num_points for _ in range(num_points)]

    for i in range(num_points):
        for j in range(num_points):
            if i == j:
                similarity_matrix[i][j] = 1.0
            else:
                similarity_matrix[i][j] = 1 / (1 + manhattan_distance(data[i], data[j]))

    return similarity_matrix
```

```python
body_manhattan_similarity_matrix = manhattan_similarity_matrix(np.array(user_profile_df_body.drop('userID',axis=1)))
```

```python
user_profile_body_similarity_df = pd.DataFrame(body_manhattan_similarity_matrix, index=user_profile_df_body['userID'], columns=user_profile_df_body['userID'])
```

```python
user_cuisines = user_cuisine_df.groupby('userID')['Rcuisine'].apply(set)
```

```python
num_users = len(user_cuisines)
similarity_matrix = [[0] * num_users for _ in range(num_users)]
```

```python
for i, user1 in enumerate(user_cuisines.index):
    for j, user2 in enumerate(user_cuisines.index):
        if i == j:
            similarity_matrix[i][j] = 1.0
        else:
            similarity_matrix[i][j] = jaccard_similarity(user_cuisines[user1], user_cuisines[user2])
```

```python
user_cuisine_similarity_df = pd.DataFrame(similarity_matrix, index=user_profile_df_temp['userID'], columns=user_profile_df_temp['userID'])
```

```python
user_cuisine_similarity_df
```

```python
pf_w = 0.1
cui_w = 0.7
loc_w = 0.1
bod_w = 0.1
```

```python
user_similarity_df = cui_w*user_cuisine_similarity_df + pf_w*user_profile_similarity_df + loc_w*user_profile_location_similarity_df + bod_w*user_profile_body_similarity_df
```

```python
user_similarity_df
```

```python
user_cuisine_matrix = user_cuisine_df.pivot_table(index='userID', columns='Rcuisine', aggfunc='size', fill_value=0)
```

```python
user_cuisine_matrix
```

```python
def get_top_k_cuisines_for_user(user_id, k=5):
    # Find similar users to the target user
    similar_users = user_similarity_df[user_id].sort_values(ascending=False)[1:k+5].index

    # Get the cuisines preferred by similar users
    top_k_cuisines = user_cuisine_matrix.loc[similar_users].sum().sort_values(ascending=False)[:k].index

    return top_k_cuisines
```

```python
user_id_to_recommend = 'U1016'  # Replace with the user ID you want to recommend cuisines for
top_k_cuisines = get_top_k_cuisines_for_user(user_id_to_recommend, k=5)
print(f"Top-K cuisines for user {user_id_to_recommend}:")
print(top_k_cuisines)
```
